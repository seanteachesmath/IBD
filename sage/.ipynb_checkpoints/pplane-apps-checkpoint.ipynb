```python
var('B C');
r = 2;
K = 10;
p = 0.5;
epsilon = 0.1;
m = 0.35;
xmax = 20;
ymax = 5;
xl = 0.025*xmax;
yl = 0.025*ymax;
dB = r*B*(1-B/K) - p*B*C
dC = epsilon*p*B*C - m*C
b0 = r/p*(1-B/K)
c0 = m/(epsilon*p)

bnull = plot((r/p)*(1-x/K),x,0,xmax, thickness = 2, ymin = -1, ymax = ymax, axes_labels = ["$B$", "$C$"], color = "red") + line([(0,-2), (0, ymax)], thickness = 2, color = "red")
bnull
cnull = plot(0,x,0,xmax, thickness = 2, ymin = -1, ymax = ymax) + line([(m/(epsilon*p),-2), (m/(epsilon*p), ymax)], thickness = 2)
plt = bnull + cnull

for xi in srange(0.5, xmax, (xmax/20)):
    y1 = b0.subs(B = xi)    
    s1 = sign(dC).subs(B = xi, C = y1)
    yhead = (xi, y1 + (s1*yl))
    ytail = (xi, y1 - (s1*yl))
    plt = plt + arrow(ytail, yhead, width = 0.5, color = "black", arrowsize=2.5)

    y1 = 0    
    s1 = sign(dB).subs(B = xi, C = y1)
    xhead = (xi + (s1*xl), y1)
    xtail = (xi - (s1*xl), y1)
    plt = plt + arrow(xtail, xhead, width = 0.5, color = "black", arrowsize=2.5)    

for yi in srange(0.5, ymax, (ymax/10)):
    #x1 = b0.subs(C = yi)    
    s1 = sign(dC).subs(B = 0, C = yi)
    yhead = (0, yi + (s1*yl))
    ytail = (0, yi - (s1*yl))
    plt = plt + arrow(ytail, yhead, width = 0.5, color = "black", arrowsize=2.5)

    xi = m/(epsilon*p)    
    s1 = sign(dB).subs(B = xi, C = yi)
    xhead = (xi + (s1*xl), yi)
    xtail = (xi - (s1*xl), yi)
    plt = plt + arrow(xtail, xhead, width = 0.5, color = "black", arrowsize=2.5)

for xi in srange(0.5, xmax, (xmax/19)):
    for yi in srange(0.5, ymax, (ymax/9)):
        xlen = dB.subs(B = xi, C = yi)
        if xlen - xi != 0:
            xlen = xlen/abs(xlen)
            ylen = dC.subs(B = xi, C = yi)
            m = (ylen - yi)/(xlen - xi)
            xhead = xi + xlen
            yhead = yi + m*(xhead - xi)
            plt = plt + arrow((xi,yi), (xhead,yhead), width = 0.5, color = "gray", arrowsize=2.5, arrowshorten = 10, xmin = 0, xmax = xmax)        
plt
```
