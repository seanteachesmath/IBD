```python
def table():
    print(' x & f(x) \\\\')## \\ to print one \
    for x in [-5 .. -1, 1 .. 5]:
        print(' %f & %f \\\\'%(x, sin(x)))
        ## look into printing number of digits
table()
```

```python
def sec_table():
    print('del.x & x1 & f(x1) & del.f & msec \\\\')
    f0 = f(x0)
    p = plot(f(x), x, 0, 5, color = 'black')
    for run in [2.0, 1.0, 0.1, 0.01, 0.001]:
        ## use 'run' as 'delta x'
        x1 = x0 + run
        f1 = f(x1)
        rise = f1 - f0
        msec = rise/run
        p = p + plot(f0 + msec*(x-x0), x, 0, 5)
        print(' %.3f & %.3f & %f & %f & %f \\\\'%(run, x1, f1, rise, msec))
        ## thanks stephen for the .3
    return(p)   
f(x) = exp(-x)
x0 = 1
sec_table()
```
